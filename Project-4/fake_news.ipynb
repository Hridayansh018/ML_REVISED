#### Importing the dependencies

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import re
import joblib
import string
```

#### Data Preprocessing

loading the dataset

```python
fake = pd.read_csv('./Fake.csv')
true = pd.read_csv('./True.csv')
```

```python
fake.head()
```

```python
true.head()
```

Giving Class for both true and fake news

```python
fake['class'] = 0
true['class'] = 1
```

```python
fake.head()
```

```python
true.head()
```

merging bothe the fake and true datasets

```python
data = pd.concat([fake,true], axis=0)
data.sample(15)
```

dropping the unnecessary titals

```python
data = data.drop(['title','subject','date'], axis=1)
```

```python
data.sample(10)
```

resetting the index

```python
data.reset_index(inplace=True)
```

```python
data.sample(10)
```

```python
data.drop(['index'],axis=1,inplace=True)
```

```python
data.sample(10)
```

Text cleaning

```python
def clean_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]',"",text)
    text = re.sub("\\W"," ",text)
    text = re.sub("https?:://\S+","",text)
    text = re.sub("<.*?>+","",text)
    text = re.sub("[%s]" % re.escape(string.punctuation),"",text)
    text = re.sub("\n","",text)
    text = re.sub("\w*\d\w*","",text)
    return text
```

```python
data['text'] = data['text'].apply(clean_text)
```

Train Test Split the data

```python
x=data['text']
y=data['class']
```

```python
xtrain, xtext, ytrain, ytest = train_test_split(x,y, random_state=2, test_size=0.25)
```

Vectorizing the text data

```python
vectorizer = TfidfVectorizer()
xvtrain = vectorizer.fit_transform(xtrain)
xvtest = vectorizer.transform(xtext)
```

training the model

```python
lr = LogisticRegression()
lr.fit(xvtrain,ytrain)
```

Model evealuation

```python
predictions = lr.predict(xvtest)
lr.score(xvtest,ytest)
```

final output

```python
print(classification_report(ytest,predictions))
```
